## Criando modelo de clusterização

Nessa etapa do caderno treinaremos um modelo de aprendizado não supervisionada para realizar clusterização. Usaremos o algoritmo kmeans.

### Selecionando as variáveis que serão utilizadas

```python
df.head() 
```

```python
cols_num = ['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
            'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'Days_Customer']
```

```python
df_cluster = df[cols_num]
```

```python
df_cluster.plot(kind='box', figsize = (20,4))
```

### Transformando variáveis

```python
# Para kmeans, ter uma variável com variância muito superior às outras é ruim, vamos padronizar os dados
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df_cluster_scaler = scaler.fit_transform(df_cluster)
df_cluster_scaler
```

```python
df_cluster_scaler = pd.DataFrame(df_cluster_scaler, columns=cols_num)
df_cluster_scaler
```

```python
df_cluster_scaler.plot(kind = 'box', figsize = (20,4))
```

### Rodando o algoritmo kmeans

```python
#Definindo kmeans com 4 clusteres
kmeans = KMeans(4)
```

```python
clusteres = kmeans.fit_predict(df_cluster_scaler)
```

```python
# Inertia é o valor que o kmeans tenta minizar, geralmente consiste no quadrado da distância ao centro dos clusteres
kmeans.inertia_
```

### Tentando encontrar o melhor valor de inertia variando a quantidade de clusteres

```python
ks_lst = range(2,21)
```

```python
# Biblioteca que auxilia a verificar a execução da célula
from tqdm import tqdm
```

```python
inertias_lst = []
for i in tqdm(ks_lst):
    kmeans = KMeans(i)
    kmeans.fit(df_cluster_scaler)
    inertias_lst.append(kmeans.inertia_)
```

```python
inertias_lst
```

```python
plt.plot(ks_lst, inertias_lst)
plt.xticks(ks_lst)
plt.plot(figsize=(20,4))
```

Olhando a imagem acima vemos que a maior variação de inércia vem da mudança de k=2 para k=3, vamos usar k=3. Lembre-se um bom modelo possui baixo valor de inércia mas também um baixo valor de clusteres.

### Vamos treinar o novo modelo e adicionar o valor dos clusteres nas linhas.

```python
kmeans = KMeans(3)
```

```python
clusteres = kmeans.fit_predict(df_cluster_scaler)
clusteres
```

```python
df['Clusteres'] = clusteres
```

```python
df
```
