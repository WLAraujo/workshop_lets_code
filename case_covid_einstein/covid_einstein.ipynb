```python
from sklearn.linear_model import LogisticRegression
```

Vamos agora construir um pipeline de Machine Learning que trata os valores nulos e treina um modelo.

```python
cols_obj = df.select_dtypes(include = 'O').columns[1:].tolist()
cols_obj
```

```python
x = df.drop(columns = ['SARS-Cov-2 exam result'])
y = df['SARS-Cov-2 exam result']
```

```python
x.drop(columns=cols_obj, inplace = True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, stratify = y, random_state = 42)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

Vamos tratar dos valores nulos usando o inputer que é a estratégia para tratar nulos.

```python
imputer = SimpleImputer(strategy='constant', fill_value=-99)
classifier = LogisticRegression(solver='liblinear')
pipe = Pipeline([('imputer', imputer), ('classifier', classifier)])
```

```python
pipe
```

```python
pipe.fit(x_train, y_train)
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
plot_confusion_matrix(pipe, x_test, y_test, cmap = 'Blues')
```
